## Background for Logistic Regression
Previously we have learned _linear regression_ that can map input X values into output Y. Here, both X and Y are continuous. <br>
<br>
Logistic regression is a kind of classification because the output value is not continuous anymore. The output becomes discrete that is either 0 or 1. <br>
<br>
The general idea of logistic regression is similar to linear regression that it performs some regression model to a $d*n$ dimension of input data X and output Y of dimension $n$. The difference is that logistic regression aims to get an output of Y which only belongs to 0 and 1.<br>
<br>
Therefore, our goal is to get a discrete output from a continuous input. <br>
## Deriving Logistic Function
Since the difference between classification and regression is only about the output. We can do some transformations to the well-known linear regression and derive logistic regression from it. <br>
<br>
For example, imagine we have a data that includes X as any real number and Y as 0 and 1, as shown below.<br>
<br>


If we simply perform linear regression to the data as our model, the regression will be linear and output will be continuous. The Y may even be predicted as a number that is more than the actual largest value of Y (more than 1). It is out of the range. Therefore, we need to make the output unbounded. <br>

<br>
From our mathematical background, we know that logit function is bounded. We can make use of that to find a regression model that satisfies our purposes. Logit function is good on its unboundedness, but it appears in a reversed order. That is, the input is 0 and 1 while the output is continuous. It maps X which is between 0 and 1 into Y which is from $-\infty$ to $\infty$. To make it as the input and output we want, we can map the output back to 0 and 1. <br>
<br>
To be more specific, the mathematical process is shown below. <br>
<br>
Since $P(Y)$ is $w^Tx$ in the linear regression, we have:
$$ logit(P(Y))=ln(\frac{P(Y)}{1-P(Y)})=w^Tx $$
To get rid of $ln$, the equation becomes:
$$ \frac{P(Y)}{1-P(Y)}=e^{w^Tx} $$
Therefore, we get the equation for $P(Y)$ is:
$$ P(Y)=e^{w^Tx}-P(Y)(e^{w^Tx}) $$
To make it shorter:
$$ P(Y)=\frac{e^{w^Tx}}{1+e^{w^Tx}} $$
<br>
Here, we have successfully derived a regression function that is called _logistic regression_.

## Closer Look at Logistic Regression
To formally write the logistic regression, we can change $w^Tx$ to $\beta_0+\beta_1x_1$:
$$ P(Y)=\frac{e^{\beta_0+\beta_1x_1}}{1+e^{\beta_0+\beta_1x_1}} $$
This logistic regression satisfies all of our goals:
- continuous
- non-linear
- bounded <br>

To interpret the parameters of the logistic regression fucntion, we can think about a simplified model. <br>
<br>
For a simplified model with X being only 1 dimension, the $\beta_1$ can be interpreted as the slope of the regression, and $\beta_0$ is the position of the regression. This is just like how we interpreted the parameters of linear regression.<br>
<br>
Now, we understand the basics for interpreting the logistic regression function, the problem is: how are we going to use it? <br>
<br>
The examples we used above all have the output Y being 0 and 1, but it does not always have to be equal to 0 and 1. Since we know it is a kind of classification, the output can be any two discrete values, or even categorical values. What we do is just map the output into 2 different boxes. <br>
<br>
For example, we can run a data of people's score of depression. Here the X is the score of depression and Y is the whetehr the person is happy or not. We can run a logistic regression to achieve that. The regression helps us predicting whether a person with certain score of depression is happy or not. <br>
